```python
## get_rsi.py

from tradingview_ta import TA_Handler, Interval, Exchange
from datetime import datetime
import time
import csv
import pandas as pd

# import generic api to clean csv
import little_tools as ll

start_time = time.time()

## removing 1st column 'screener' of csv screener files
def clean_csv():
    ## removing 1st column of csv screener files
    data_frame = pd.read_csv('crypto_screeners.csv')
    data_frame.pop('screener')
    data_frame.to_csv ('crypto_screeners_rd.csv', encoding='utf-8', index=False)

## filter screener that contains the values a submitted csv file with position   
def filter_position(): 
    df = pd.read_csv('positions.csv', converters={'positions': eval})
    data_frame = pd.read_csv('crypto_screeners_rd.csv')
    
    
    
    
def print_start():
    ## display start time/date
   
    # datetime object containing current date and time
    # dd/mm/YY H:M:S
    dtbegin_string = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    print("date and time begin =", dtbegin_string)

# print the execution time,end date and the number of execptions found    
def print_end():
    print ("rsi recording finsihed")
    dtend_string = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    print("date and time end =", dtend_string)
    print("execution time in  %s minutes" % ((time.time() - start_time)/60))
    print("have a good day")

# parse each crypto symbol in the screener csv file and get the RSI length 14, close
def record_rsi():    
    ## out file
    current_date = datetime.now().strftime("%d_%m_%Y")
    out_csv = "RSI_records" + "-" + current_date


    with open('crypto_screeners_rd.csv', 'r') as csv_file:
        reader = csv.reader(csv_file)
        ## remove old file
        ## create new file for rsi records
        with open(out_csv+ ".csv", 'w') as csv_file:
            writer = csv.writer(csv_file)
            rsi_infoshd = {}
            rsi_infoshd['description'] = []
            rsi_infoshd['symbol'] = []
            rsi_infoshd['RSI 1D: *length: 14 days *source: close)'] = []
            rsi_infoshd['exchange'] = []
            rsi_infoshd['date and time'] = []
            writer.writerow(rsi_infoshd)
            scrn="crypto"
            count_excp = 0
            count = 0

            ## format crypto_name= line[1], sym =line[2], ex=line[0]  
            for line in reader:
                ##crypto_name= line[2]
                ##sym =line[2]
                ##ex=line[1]
                count = count + 1
                Crypto_hdl = TA_Handler(
                symbol=line[1],
                exchange=line[0],
                screener= scrn,
                interval=Interval.INTERVAL_1_DAY,
                timeout=None
                )
                try: 
                    analysis= Crypto_hdl.get_analysis()
                except:
                    count_excp =count_excp +1
                    continue;
                else:
                    if analysis != None:
                            ##print(analysis)
                        ##print(count)
                        ##print(line)
                        rsi= analysis.indicators["RSI"]
                        ## update time
                        t_string = datetime.now().strftime("%H:%M:%S")

                            ## description/ full name,symbol,RSI, exchange, date and time
                        rsi_infos = [line[2],line[1],rsi,line[0],t_string]
                        #print(rsi_infos)
                        writer.writerow(rsi_infos)

                        del analysis
                        del Crypto_hdl
                        #if count > 10:
                        #    break;

            del writer
            del reader
            print ("number of exceptions:", count_excp)
# display time infos for start            
print_start()
# put csv from db in form for the script
clean_csv()
# collect RSI for each crypto
record_rsi()
# display time infos for end and execution time and execptions 
print_end()
```
